```python
def featureExtractor(wordList):
    #contains the output multidimensional array which contains all the pArrays
    oArray = []
    #contains the features per processed word
   
    
    fDict = {}
    #the extracted features are: word legth, count by characters, count by multi char,
    
    
    
    for word in wordList:
        pArray = []
        word = str(word)
        pArray.append(len(word)) # 1. word lenght
        
        
    
        for ch in word:
            pArray.append(ch) #2. append each character to array

        
        chcnt = 1 #character counter
        for ch in word:
            if chcnt < len(word):
                chcon = ch + word[chcnt]
                pArray.append(chcon)
                
                if chcon in fDict:
                    fDict[chcon] = fDict[chcon] + 1
                else:
                    fDict[chcon] = 1
                chcnt +=1
        
        oArray.append(pArray)
        
    
    
    oArray.append(fDict)
    
    return oArray


oArray = featureExtractor(['Vladislav', 'Julia', 'Angelina'])
print(oArray)


```

```python
nameList = dt[(dt.host_name_multstr == 1) & (dt.host_name_len < 10) & (dt.host_gender != 'GR')].host_name.to_list()
```

```python
bigArray = featureExtractor(nameList)
bigDict = bigArray[-1]
type(bigDict)
```
