```python
import pandas as pd
```

```python
airdt = pd.read_csv('AB_NYC_2019.csv')
```

```python
airdt.head()
```

```python
def conand(dt):
    word = str(dt).lower()
    if word.__contains__(' and '):
        return True
    elif word.__contains__('&'):
        return True
    elif word.__contains__(' or '):
        return True
    else:
        return False
```

```python
airdt['conAnd'] = airdt.host_name.apply(conand)
```

```python
airdt[airdt.conAnd == True].loc[1:,['host_name', 'conAnd']]
```

```python
airdt['host_gender'] = airdt.conAnd.map({True:'GR',False:'0'})
```

```python
airdt[airdt.conAnd == False].loc[1:,['host_name', 'conAnd', 'host_gender']]
```

```python
def lastL(dt):
    name = str(dt)
    return name[-1]
```

```python
airdt['host_lastL'] = airdt[airdt.conAnd == False].host_name.apply(lastL)
```

```python
airdt[airdt.conAnd == False].loc[1:,['host_name', 'conAnd',  'host_lastL', 'host_gender']]
```

```python
from genderize import Genderize
#response = Genderize().get(['James'])


nameList = airdt[airdt.conAnd == False].host_name.tolist()
#nameList = [1,2,3,4,5,6,7,8,9,0,10,11,12,143,1,14,144,14,34,2354,35,35,46]
lastBatchNum = 0
response = []
for batch in range(0,1000,10):
    listSelection = nameList[lastBatchNum:batch]
    response = Genderize().get(listSelection)
    response.append(response)
    lastBatchNum = batch   
    
print('re' + str(response))


```

```python
f = open('gender.csv', 'w+')
f.write('name,gender,probability,count')
f.close()
```
