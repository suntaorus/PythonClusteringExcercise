```python
import pandas as pd
```

```python
airdt = pd.read_csv('AB_NYC_2019.csv')
```

```python
airdt.head()
airdt.fillna('NaN')
```

```python
def conand(dt):
    word = str(dt).lower()
    if word.__contains__(' and '):
        return True
    elif word.__contains__('&'):
        return True
    elif word.__contains__(' or '):
        return True
    else:
        return False
```

```python
airdt['conAnd'] = airdt.host_name.apply(conand)
```

```python
airdt[airdt.conAnd == True].loc[1:,['host_name', 'conAnd']]
```

```python
airdt['host_gender'] = airdt.conAnd.map({True:'GR',False:'0'})
```

```python
airdt[airdt.conAnd == False].loc[1:,['host_name', 'conAnd', 'host_gender']]
```

```python
def lastL(dt):
    name = str(dt)
    return name[-1]
```

```python
airdt['host_lastL'] = airdt[airdt.conAnd == False].host_name.apply(lastL)
```
